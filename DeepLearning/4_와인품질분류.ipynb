# 레드와인의 화학적 성분과 등급을 기록한 데이터
- 주소 : https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv

```python
import pandas as pd

redwine = pd.read_csv('winequality-red.csv', sep=';')
redwine.head()
```

```python
from sklearn.model_selection import train_test_split

redwine_X = redwine.iloc[:, :-1].to_numpy()
redwine_y = pd.get_dummies(redwine.iloc[:,-1]).to_numpy()

#train-test
train_X, test_X, train_y, test_y = train_test_split(redwine_X, redwine_y, test_size=0.3, random_state=1)
train_X.shape, train_y.shape
```

# 문제 정의 : 와인의 품질 속성을 분류하기
## 모형 설계

Sequential()
- 레이어1 : 뉴런 50, 활성화 함수 relu
- 레이어2 : 뉴런 50, 활성화 함수 relu
- 레이어3 : 뉴런 30, 활성화 함수 relu
- 레이어4 : 출력 6, 활성화 함수 softmax

compile()
- 최적화 : adam
- 손실함수 : 'categorical_crossentropy'
- 평가지표 : 'accuracy'

```python

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
# 1. 모델 정의
model = Sequential()
model.add(Input(shape=(11,)))  #입력층 11개 피처 지정
model.add(Dense(50, activation='relu')) #은닉층 레이어1
model.add(Dense(50, activation='relu')) #레이어2
model.add(Dense(30, activation='relu')) #레이어3
model.add(Dense(6, activation='softmax')) #출력층 레이어4
model
```

```python
# 2. 모델 설정 -컴파일
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model
```

```python
# 3. 모델 훈련
model.fit( train_X, train_y, epochs=200, batch_size=50, verbose=1)
```

```python
# 4. 평가
import numpy as np

pred = np.argmax(model.predict( test_X ), axis=1)
pred # 예측값
```

- 모형 구조를 변경
- 최적화 함수 바꾸기
- 에포크 횟수 늘리기

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

# 데이터 로드
redwine = pd.read_csv('winequality-red.csv', sep=';')

# 특성과 타겟 분리
redwine_X = redwine.iloc[:, :-1].to_numpy()
redwine_y = pd.get_dummies(redwine.iloc[:,-1]).to_numpy()

# 훈련 세트와 테스트 세트 분리
train_X, test_X, train_y, test_y = train_test_split(redwine_X, redwine_y, test_size=0.3, random_state=1)

# 모델 정의
model = Sequential([
    Input(shape=(11,)),
    Dense(128, activation='relu'),
    BatchNormalization(),
    Dropout(0.3),
    Dense(64, activation='relu'),
    BatchNormalization(),
    Dropout(0.3),
    Dense(32, activation='relu'),
    BatchNormalization(),
    Dense(6, activation='softmax')
])

# 옵티마이저 설정
optimizer = Adam(learning_rate=0.001)

# 모델 컴파일
model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])

# 콜백 함수 정의
early_stopping = EarlyStopping(patience=20, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(factor=0.2, patience=5, min_lr=0.00001)

# 모델 학습
history = model.fit(
    train_X, train_y,
    epochs=500,
    batch_size=32,
    validation_split=0.2,
    callbacks=[early_stopping, reduce_lr],
    verbose=1
)

# 모델 예측
pred = np.argmax(model.predict(test_X), axis=1)

# 혼동 행렬 생성
confusion_matrix = pd.crosstab(np.argmax(test_y, axis=1), pred, margins=True)

# 모델 평가
loss, accuracy = model.evaluate(test_X, test_y)
print(f"Loss: {loss}, Accuracy: {accuracy}")

# 모델 요약
model.summary()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.optimizers import Adam

# 데이터 로드
redwine = pd.read_csv('winequality-red.csv', sep=';')

# 특성과 타겟 분리
redwine_X = redwine.iloc[:, :-1].to_numpy()
redwine_y = pd.get_dummies(redwine.iloc[:, -1]).to_numpy()

# 훈련 세트와 테스트 세트 분리
train_X, test_X, train_y, test_y = train_test_split(redwine_X, redwine_y, test_size=0.3, random_state=1)

# 데이터 정규화
scaler = StandardScaler()
train_X = scaler.fit_transform(train_X)
test_X = scaler.transform(test_X)

# 모델 정의
model = Sequential([
    Input(shape=(11,)),
    Dense(128, activation='relu', kernel_initializer='he_normal'),
    Dense(64, activation='relu', kernel_initializer='he_normal'),
    Dense(32, activation='relu', kernel_initializer='he_normal'),
    Dense(6, activation='softmax')
])

# 모델 컴파일 (학습률 0.001로 조정)
optimizer = Adam(learning_rate=0.001)
model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])

# 모델 학습
model.fit(train_X, train_y, epochs=500, batch_size=32, verbose=1)

# 모델 예측
pred = np.argmax(model.predict(test_X), axis=1)

# 혼동 행렬 생성
confusion_matrix = pd.crosstab(np.argmax(test_y, axis=1), pred, margins=True)

# 모델 평가
loss, accuracy = model.evaluate(test_X, test_y)
print(f"Loss: {loss}, Accuracy: {accuracy}")

# 모델 요약
model.summary()
```
