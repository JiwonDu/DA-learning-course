```python
host_ip = "13.208.159.5"
user_id = "spark_user"
user_password = "1234"
```

```python
import os
os.environ["PYSPARK_SUBMIT_ARGS"] = "--jars /usr/local/lib/mysql-connector-java-5.1.49-bin.jar pyspark-shell"
```

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("mysqlconnect").getOrCreate()
```

```python
mysql_url = f"jdbc:mysql://{host_ip}:3306/my_db"
dept_df = spark.read.format("jdbc").options(
    url=mysql_url,
    driver="com.mysql.jdbc.Driver",
    dbtable="DEPT",
    user=user_id,
    password=user_password
).load()
```

```python
dept_df.show()
```

```python
# EMP, SALGRADE

# EMP 테이블 가져오기
emp_df = spark.read.format("jdbc").options(
    url=mysql_url,
    driver="com.mysql.jdbc.Driver",
    dbtable="EMP",
    user=user_id,
    password=user_password
).load()

# SALGRADE 테이블 가져오기
salgrade_df = spark.read.format("jdbc").options(
    url=mysql_url,
    driver="com.mysql.jdbc.Driver",
    dbtable="salgrade",
    user=user_id,
    password=user_password
).load()
```

```python
emp_df.show()
```

```python
salgrade_df.show()
```

```python
# sql()
# DEPT 데이터프레임 템프뷰 등록
dept_df.createOrReplaceTempView("DEPT")

# EMP 데이터프레임 템프뷰 등록
emp_df.createOrReplaceTempView("EMP")

# SALGRADE 데이터프레임 템프뷰 등록
salgrade_df.createOrReplaceTempView("salgrade")
```

```python
# Spark SQL을 사용하여 DEPT 테이블에서 쿼리 실행
result_df = spark.sql("SELECT * FROM DEPT")

# 결과 출력
result_df.show()

# EMP 테이블에서 쿼리 실행
result_emp_df = spark.sql("SELECT * FROM EMP")

# 결과 출력
result_emp_df.show()

# SALGRADE 테이블에서 쿼리 실행
result_salgrade_df = spark.sql("SELECT * FROM salgrade")

# 결과 출력
result_salgrade_df.show()
```

```python
# DEPT, EMP, SALGRADE 테이블을 한 번에 조인
full_join = dept_df.join(emp_df, dept_df.DEPTNO == emp_df.DEPTNO) \
                   .join(salgrade_df, (emp_df.SAL >= salgrade_df.losal) & (emp_df.SAL <= salgrade_df.hisal))

# 결과 출력
full_join.show()
```
