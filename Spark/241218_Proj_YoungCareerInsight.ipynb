````python
# SQL 쿼리로 데이터 불러오기
satisfaction_data = pd.read_sql('SELECT * FROM satisfaction', con=engine)
values_data = pd.read_sql('SELECT * FROM job_value', con=engine)

# 만족도 항목 리스트
satisfaction_columns = [
    'w01Ba30', 'w01Ba30_1_1', 'w01Ba30_1_2', 'w01Ba30_1_3', 'w01Ba30_1_4', 
    'w01Ba30_1_5', 'w01Ba30_1_6', 'w01Ba30_1_7', 'w01Ba30_1_8', 'w01Ba30_1_9', 
    'w01Ba30_1_10', 'w01Ba30_1_11', 'w01Ba30_1_12'
]

# 분포를 저장할 데이터프레임 생성
satisfaction_distribution = pd.DataFrame()

# 각 항목의 분포 계산
for column in satisfaction_columns:
    distribution = satisfaction_data[column].value_counts().sort_index()
    satisfaction_distribution[column] = distribution

# 결측치를 0으로 대체
satisfaction_distribution = satisfaction_distribution.fillna(0)
print(satisfaction_distribution)
````

````python
import matplotlib.pyplot as plt
import numpy as np

# 히스토그램을 그리기 위한 설정
fig, axes = plt.subplots(nrows=3, ncols=5, figsize=(25, 15))  # 3x5 그리드 설정
axes = axes.flatten()

# 모든 데이터의 최대 빈도수 계산
max_frequency = 0
for column in satisfaction_columns:
    counts, _ = np.histogram(satisfaction_data[column], bins=np.arange(1, 7))
    max_frequency = max(max_frequency, max(counts))

# 각 만족도 항목에 대해 히스토그램 생성 및 강조
for i, column in enumerate(satisfaction_columns):
    data = satisfaction_data[column]
    counts, bins, patches = axes[i].hist(data, bins=np.arange(1, 7), edgecolor='black')
    axes[i].set_title(column)
    axes[i].set_xlabel('Score')
    axes[i].set_ylabel('Frequency')
    axes[i].set_xticks([1, 2, 3, 4, 5])  # x축 간격 설정
    axes[i].set_ylim(0, max_frequency * 1.1)  # y축 범위 통일 (여유 공간 10% 추가)
    
    # 중앙값 표시
    median_value = np.median(data)
    axes[i].axvline(median_value, color='blue', linestyle='dashed', linewidth=2, label=f'Median: {median_value}')
    
    # 가장 많이 표시된 데이터를 강조 (최빈값)
    mode_value = data.mode()[0]
    for count, patch in zip(counts, patches):
        if count == max(counts):
            patch.set_facecolor('red')  # 강조 색상 설정
    
    axes[i].legend()

# 남은 subplot 제거
for j in range(len(satisfaction_columns), len(axes)):
    fig.delaxes(axes[j])

# 그래프 간 간격 조정
plt.tight_layout()
plt.show()
````

````python
from pyspark.ml.regression import RandomForestRegressor, DecisionTreeRegressor
from pyspark.ml.classification import LogisticRegression
from pyspark.ml.evaluation import RegressionEvaluator

# 평가 지표 함수
def evaluate_model(predictions, labelCol='w01Ba30'):
    evaluator_rmse = RegressionEvaluator(labelCol=labelCol, predictionCol='prediction', metricName='rmse')
    evaluator_r2 = RegressionEvaluator(labelCol=labelCol, predictionCol='prediction', metricName='r2')
    rmse = evaluator_rmse.evaluate(predictions)
    r2 = evaluator_r2.evaluate(predictions)
    return rmse, r2

# 모델 리스트
models = {
    "RandomForest": RandomForestRegressor(featuresCol='scaledFeatures', labelCol='w01Ba30'),
    "DecisionTree": DecisionTreeRegressor(featuresCol='scaledFeatures', labelCol='w01Ba30')
}

# 모델 학습 및 평가
results = {}
for model_name, model in models.items():
    print(f"Training {model_name}...")
    trained_model = model.fit(train_data)
    predictions = trained_model.transform(test_data)
    rmse, r2 = evaluate_model(predictions)
    results[model_name] = {'RMSE': rmse, 'R2': r2}

# 결과 출력
for model_name, metrics in results.items():
    print(f"{model_name} - RMSE: {metrics['RMSE']}, R2: {metrics['R2']}")
````

 → 서열척도에 적합하면서도 pyspark에서 지원하는 모델 중 비교 분석함.

````python
from pyspark.ml.clustering import KMeans
from pyspark.ml.evaluation import ClusteringEvaluator

# 실루엣 점수를 저장할 리스트
silhouette_scores = []

# 클러스터 개수를 2부터 10까지 테스트
for k in range(2, 11):
    kmeans = KMeans(k=k, seed=1234, featuresCol='scaledFeatures', predictionCol='cluster')
    kmeans_model = kmeans.fit(scaled_data)
    predictions = kmeans_model.transform(scaled_data)
    
    evaluator = ClusteringEvaluator(featuresCol='scaledFeatures', predictionCol='cluster', metricName='silhouette')
    silhouette_score = evaluator.evaluate(predictions)
    silhouette_scores.append((k, silhouette_score))
    print(f"Number of Clusters = {k}, Silhouette Score = {silhouette_score}")

# 최적의 클러스터 개수 찾기
optimal_k = max(silhouette_scores, key=lambda x: x[1])[0]
print(f"Optimal number of clusters: {optimal_k}")
````

````python
# 최적의 클러스터 개수를 사용하여 K-means 모델 학습
kmeans = KMeans(k=optimal_k, seed=1234, featuresCol='scaledFeatures', predictionCol='cluster')
kmeans_model = kmeans.fit(scaled_data)

# 예측
clustered_data = kmeans_model.transform(scaled_data)
````

````python
# 평가 지표 함수
def evaluate_model(predictions, labelCol='w01Ba30'):
    evaluator_rmse = RegressionEvaluator(labelCol=labelCol, predictionCol='prediction', metricName='rmse')
    evaluator_r2 = RegressionEvaluator(labelCol=labelCol, predictionCol='prediction', metricName='r2')
    rmse = evaluator_rmse.evaluate(predictions)
    r2 = evaluator_r2.evaluate(predictions)
    return rmse, r2

# 모델 리스트
models = {
    "RandomForest": RandomForestRegressor(featuresCol='scaledFeatures', labelCol='w01Ba30'),
    "GBTRegressor": GBTRegressor(featuresCol='scaledFeatures', labelCol='w01Ba30'),
    "DecisionTree": DecisionTreeRegressor(featuresCol='scaledFeatures', labelCol='w01Ba30'),
    "GeneralizedLinear": GeneralizedLinearRegression(featuresCol='scaledFeatures', labelCol='w01Ba30', regParam=0.01),
    "LinearRegression": LinearRegression(featuresCol='scaledFeatures', labelCol='w01Ba30', regParam=0.01)
}

# 모델 학습 및 평가
results = {}
for model_name, model in models.items():
    print(f"Training {model_name}...")
    trained_model = model.fit(train_data)
    predictions = trained_model.transform(test_data)
    rmse, r2 = evaluate_model(predictions)
    results[model_name] = {'RMSE': rmse, 'R2': r2}

# 결과 출력
for model_name, metrics in results.items():
    print(f"{model_name} - RMSE: {metrics['RMSE']}, R2: {metrics['R2']}")
````

````python
from pyspark.sql.functions import col, mean
from pyspark.sql.window import Window
from pyspark.ml.feature import VectorAssembler, StandardScaler

# 필요한 라이브러리 불러오기
selected_columns = [
    'w01Ba30_1_1', 'w01Ba30_1_2', 'w01Ba30_1_3', 'w01Ba30_1_4', 
    'w01Ba30_1_5', 'w01Ba30_1_6', 'w01Ba30_1_7', 'w01Ba30_1_8', 
    'w01Ba30_1_9', 'w01Ba30_1_10', 'w01Ba30_1_11', 'w01Ba30_1_12'
]

# Window 객체 정의 (예: 'sampid'를 기준으로 파티셔닝)
windowSpec = Window.partitionBy('sampid')  # 'sampid' 컬럼을 기준으로 파티셔닝

# 평균 만족도 계산
for col_name in selected_columns:
    cleaned_data = cleaned_data.withColumn(f"avg_{col_name}", mean(col(col_name)).over(windowSpec))

# 피처 벡터 생성
assembler = VectorAssembler(inputCols=[f"avg_{col}" for col in selected_columns], outputCol='features')
data = assembler.transform(cleaned_data)
scaler = StandardScaler(inputCol='features', outputCol='scaledFeatures')
scaler_model = scaler.fit(data)
scaled_data = scaler_model.transform(data)
````

````python
from pyspark.ml.clustering import KMeans

# K-means 모델 학습
kmeans = KMeans(k=4, seed=1234, featuresCol='scaledFeatures', predictionCol='cluster')
kmeans_model = kmeans.fit(scaled_data)

# 예측
clustered_data = kmeans_model.transform(scaled_data)
````

````python
# 클러스터 레이블과 주요 특성 조회
clustered_data.select('scaledFeatures', 'cluster').show(10)
````

````python
# 클러스터별 데이터 포인트 수 계산
cluster_counts = clustered_data.groupBy('cluster').count()

# 결과 표시
cluster_counts.show()
````

````python
from pyspark.ml.clustering import KMeans
from pyspark.ml.evaluation import ClusteringEvaluator

# 최적의 클러스터 수 찾기
best_k = 2
best_silhouette = -1
silhouette_scores = []

for k in range(2, 11):  # 2에서 10까지 클러스터 수 시도
    kmeans = KMeans(k=k, seed=1234, featuresCol='scaledFeatures', predictionCol='cluster')
    kmeans_model = kmeans.fit(scaled_data)
    clustered_data = kmeans_model.transform(scaled_data)
    
    evaluator = ClusteringEvaluator(predictionCol='cluster')
    silhouette = evaluator.evaluate(clustered_data)
    silhouette_scores.append((k, silhouette))
    
    if silhouette > best_silhouette:
        best_silhouette = silhouette
        best_k = k

print(f"Optimal number of clusters: {best_k} with silhouette score: {best_silhouette}")
print("Silhouette scores for different cluster numbers:", silhouette_scores)
````
