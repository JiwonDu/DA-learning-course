## 새로운 데이터 셋 로드
'tips' 데이터 셋, 팁, 청구서 전체금액, 일행 수, 요일, 시간, 흡연여부, 성별

```python
tips = sns.load_dataset('tips')
tips.head()
```

## 1. 일변량 그래프

### 1. 히스토그램
- 구간의 *빈도수*를 시각화하는 그래프, x의 구간을 나누어 y는 빈도수

```python
# 연결형 변수
fig = plt.figure()
axes = fig.add_subplot(1,1,1)
axes.hist(data = tips, x ='total_bill', bins=20) # bin
fig.suptitle('tips 데이터로 보는 일변량 그래프 - 히스토그램')
plt.show()
```

### 2. 막대 그래프

```python
# 이산형
fig = plt.figure()
axes = fig.add_subplot(1,1,1)
sns.countplot(data = tips, x ='day')
fig.suptitle('tips 데이터로 보는 일변량 그래프 - 막대그래프')
plt.show()
```

### 3. 상자그림

```python
# 연속형 변수 / 4분위수, 최대, 최소, 중앙값 표시
fig = plt.figure()
axes = fig.add_subplot(1,1,1)
axes.boxplot(data = tips, x ='total_bill')
fig.suptitle('tips 데이터로 보는 일변량 그래프- boxplot')
plt.show()
```

cf. 이상치 : 4분범위의 1.5배 밖에 있는 데이터

## 2. 이변량 그래프

#### 1. 산점도(scatter plot)

```python
fig = plt.figure()
axes = fig.add_subplot(1,1,1)
axes.scatter(tips['total_bill'],tips['tip']) #총 지불금액과 팁의 관계 : 우상향 → '상관관계'가 있다.
plt.show()
```

### 2. 상자그림

```python
fig = plt.figure()
axes = fig.add_subplot(1,1,1)

axes.boxplot(
    x=[ tips[tips['sex'] == 'Female']['tip'],
        tips[tips['sex'] == 'Male']['tip'] ],
    labels = ['Female', 'Male']
)
plt.show()
```

## 3. 다변량 그래프

### 1. 산점도

```python
#map()함수에 사용
colors = {"Female": "#f1a340", "Male": "#998ec3"} # 성별에 따른 색상을 지정, 여성은 주황색, 남성은 보라색 
#16진수 코드는 https://www.w3schools.com/colors/colors_picker.asp 에서 확인 가능

scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1, 1, 1)

axes1.scatter(data=tips,
              x='total_bill',
              y='tip',
              s=tips['size']**2*10,# size를 제곱하여 점의 크기를 키움, s는 점의 크기를 나타내는 매개변수
              c=tips['sex'].map(colors),# map()은 Series의 각 요소에 대해 특정 함수를 적용할 때 사용, c는 점의 색을 나타내는 매개변수, colors 딕셔너리에 따라 여성은 주황색, 남성은 보라색
              alpha=0.5) # alpha는 점의 투명도를 나타내는 매개변수

axes1.set_title('Colored by Sex and Sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')

scatter_plot.suptitle('Total Bill vs Tip')

plt.show()
```

# 3. Seaborn 라이브러리

```python
import seaborn as sns
import pandas as pd

sns.get_dataset_names()
```

## 1. 일변량 그래프

#### 1. 히스토그램

```python
sns.histplot(data=tips, x='total_bill')
plt.show()
```

### 2. 밀도 분포

```python
sns.kdeplot(data=tips, x ='total_bill')
plt.show()
```

### 3. 러그 그래프

```python
sns.rugplot(data=tips, x='total_bill')
sns.histplot(data=tips, x='total_bill')
plt.show()
```

### 4. 막대 그래프

```python
cnt, ax=plt.subplots()

sns.countplot(data=tips, x='day', palette='viridis')

ax.set_title('요일별 건수')
plt.show()
# 색상 팔레트 : https://seaborn.pydata.org/tutorial/color_palettes.html
```

## 2. 이변량 그래프

### 1. 산점도

```python
# seaborn의 장점 : 그래프를 그리기 쉽다.
cnt, ax=plt.subplots()

sns.scatterplot(data=tips, x='total_bill', y = 'tip', ax=ax, hue='sex', palette='viridis')

ax.set_title('지불금액과 팁의 관계')
plt.show()
```

### 2. reg plot

```python
cnt, ax=plt.subplots()

sns.regplot(data=tips, x='total_bill', y = 'tip', ax=ax)

ax.set_title('지불금액과 팁의 관계')
plt.show()
```

### 3. joint plot

```python
joint = sns.jointplot(data=tips, x='total_bill', y = 'tip', ax=ax)
joint.set_axis_labels(xlabel='TotalBill', ylabel='Tip')
joint.figure.suptitle('TotalBill과 Tips의 조인트 차트')
plt.show()
```

### 4. Ked plot

```python
kde, ax = plt.subplots()
sns.kdeplot( data=tips, x='total_bill', y='tip', fill=True, ax=ax)
ax.set_title("KDE Plot")
plt.show()
```

### 5. 막대 그래프

```python
bar, ax = plt.subplots()
sns.barplot( data=tips, x='time', y='total_bill', estimator=np.mean, ax=ax)
ax.set_title("bar Plot")
plt.show()
```

### 6. 상자 그림

```python
box, ax = plt.subplots()
sns.boxplot( data=tips, x='time', y='total_bill', ax=ax)
ax.set_title("box Plot")
plt.show()
```

### 7. violin plot

```python
violin, ax = plt.subplots()
sns.violinplot( data=tips, x='time', y='total_bill', ax=ax)
ax.set_title("violin Plot")
plt.show()

# 굳이 이상치로 표현하지 않음. boxplot보다 더 많은 정보!
# seaborn은 다변량이라 표현하지 않음. hue를 주면 됨.
```

## 3. pairplot
